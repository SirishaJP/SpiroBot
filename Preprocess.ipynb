```python
import cv2 as cv
import cv2.cv2
import numpy as np
import os
import random
import matplotlib.pyplot as plt
```

```python

BASE_FOLDER_PATH = f'Medical'
ANNOTATION_FOLDER_PATH = f'Label Medical'
```

```python
def read_images(folder_path=None, as_grey=0):
    list_of_images = []
    for image in sorted(os.listdir(folder_path)):
        image_path = os.path.join(folder_path, image)
        image = cv.imread(image_path, as_grey)
        list_of_images.append(image)

    return list_of_images
```

```python

def resize_images(image_list=None, width=500, height=500):
    image_stack = []
    for image in image_list:
        # Change interpolation to your choice
        image = cv.resize(src=image, dsize=(height, width), interpolation=cv2.cv2.INTER_AREA)
        image_stack.append(image)
    return image_stack
```

```python
def check_images_consistency(base_list, mask_list):
    random_number = random.randint(0, len(base_list) - 1)
    base_image = base_list[random_number]
    mask_image = mask_list[random_number]

    plt.imshow(base_image,cmap='gray')
    plt.imshow(mask_image,cmap='hot',alpha=0.4)
    plt.show()
```

```python
if __name__ == '__main__':
    # read base images
    base_images = read_images(BASE_FOLDER_PATH)
    mask_images = read_images(ANNOTATION_FOLDER_PATH)

    # Resize as required
    base_images = resize_images(image_list=base_images, width=500, height=500)
    mask_images = resize_images(image_list=mask_images, width=500, height=500)

    # Convert to numpy array
    base_images = np.array(base_images)
    mask_images = np.array(mask_images)

    # Format of shape - (number of images,height,width)
    print(f'Shape of Base : {base_images.shape}')
    print(f'Shape of Mask : {mask_images.shape}')
```
